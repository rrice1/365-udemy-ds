```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv('TryingPropAgeByAvgIndAgain.csv')
```

```python
data
```

```python
y=data['PolicyInsuredPropertyAge']
x1=data['Avg. Corrected Incurred Indemnity']
```

```python
plt.scatter(x1,y)
plt.xlabel('indemnity',fontsize=20)
plt.ylabel('age',fontsize=20)
plt.show()
```

```python
x=sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
yhat = -0.0008*x1+51.2783
fig = plt.plot(x1,yhat, lw=4, c='orange', label ='regression line')
plt.xlabel('incurred indemnity', fontsize=20)
plt.ylabel('property age', fontsize=20)
plt.show()
```
