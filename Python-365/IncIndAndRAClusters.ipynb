```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set the styles to Seaborn
sns.set()
# Import the KMeans module so we can perform k-means clustering with sklearn
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('ClaimsWithIndemnityAndRA.csv')
data
```

```python
plt.scatter(data['PolicyInsuredPropertyRoofAge'],data['Corrected Incurred Indemnity'])
plt.show
```

```python
x = data.iloc[:,1:3]
x
kmeans = KMeans(3)
```

```python
kmeans.fit(x)
```

```python
identified_clusters=kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['Corrected Incurred Indemnity'],data_with_clusters['PolicyInsuredPropertyRoofAge'],c=data_with_clusters['Cluster'],cmap='rainbow')
plt.show()
```

```python
kmeans.inertia_
```

```python
wcss = []

for i in range (1,10):
    kmeans = KMeans(i)
    kmeans.fit(x)
    wcss_iter = kmeans.inertia_
    wcss.append(wcss_iter)
```
