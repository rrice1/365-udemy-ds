```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set the styles to Seaborn
sns.set()
# Import the KMeans module so we can perform k-means clustering with sklearn
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv('TryingPropAgeByAvgIndAgain.csv')
data
```

```python
plt.scatter(data['PolicyInsuredPropertyAge'],data['Avg. Corrected Incurred Indemnity'])
plt.show
```

```python
kmeans = KMeans(4)
```

```python
kmeans.fit(data)
```

```python
identified_clusters=kmeans.fit_predict(data)
identified_clusters
```

```python
data_with_clusters = data.copy()
data_with_clusters['Cluster'] = identified_clusters
data_with_clusters
```

```python
plt.scatter(data_with_clusters['PolicyInsuredPropertyAge'],data_with_clusters['Avg. Corrected Incurred Indemnity'],c=data_with_clusters['Cluster'],cmap='rainbow')
plt.show()
```
